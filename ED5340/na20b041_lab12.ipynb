```python
# Question 1
import numpy as np
import pandas as pd

def forward(x,w,b): # forward prop
    k=np.dot(np.array(w),np.array(x).T)
    z=np.add(k,np.array(b).T)
    a=1/(1+np.exp(-z))
    return a.T,z

def backprob(x,y,a1,a2,w1,w2,b1,b2,z1):   # back prop
    
    dz2=np.subtract(a2.T,np.array(y))
    dw2=np.dot(dz2,a1)/len(a1)
    db2=np.sum(dz2)/len(a1)
    d1=np.dot(np.array(w2).T,dz2)
    d2=a1.T
    dz1=np.multiply(d1,d2)
    dw1=np.dot(dz1,x)/len(x)
    db1=dz1/len(x)
    return dw1,db1,dw2,db2,dz1

def update(dw1,db1,dw2,db2,w1,b1,w2,b2):  # update steps
    a=0.0005
    w2=w2-a*dw2
    b2=b2-a*db2
    w1=w1-a*dw1
    b1=b1-a*db1.T

    return w1,b1,w2,b2


def Neural(x,y,w1,w2,b1,b2):  # Neural net

    i=0
    si=0.00001
    dz1=1
    while(i<11400):
        i+=1
        a1,z1=forward(x,w1,b1)
        a2,z2=forward(a1,w2,b2)
        dw1,db1,dw2,db2,dz1=backprob(x,y,a1,a2,w1,w2,b1,b2,z1)
        w1,b1,w2,b2=update(dw1,db1,dw2,db2,w1,b1,w2,b2)


    return w1,b1,w2,b2

df=pd.read_csv(r"Logistic_regression_ls.csv")
df=df.to_dict('list')

x1=df['x1']
x2=df['x2']
y=df['label']

X1=[[x1[i],x2[i]] for i in range(int(len(x1)*0.8))]
X2=[[x1[-i],x2[-i]] for i in range(int(len(x1)*0.2))]
y1=[y[i] for i in range(int(len(x1)*0.8))]
y2=[y[-i] for i in range(int(len(x1)*0.2))]

b1=[[1,2,3,3,0,5,8] for i in range(len(X1))]
b2=[5]
w1=[[-1,-1],[2,3],[1,1],[2,1],[0,0],[7,6],[1,1]]
w2=[[0,1,0,1,0,1,0]]

w1,b1,w2,b2=Neural(X1,y1,w1,w2,b1,b2)  #Train


#Test data
b1=[b1[0] for i in range(len(X2))]
a1,z=forward(X2,w1,b1)
a2,z=forward(a1,w2,b2)
p=[]
for i in range(len(a2)):
    if a2[i]>=0.5:
        p.append(1)
    else:
        p.append(0)
tp=fp=tn=fn=0
for i in range(len(p)):
        if p[i]==y2[i] and p[i]==1:
            tp+=1
        elif p[i]==y2[i] and p[i]==0:
            tn+=1
        elif p[i]!=y2[i] and p[i]==1:
            fp+=1
        elif p[i]!=y2[i] and p[i]==0:
            fn+=1

pre=tp/(tp+fp)
rec=tp/(tp+fn)
f1=pre*rec/(pre+rec)
acc=(tp+tn)/(tp+fp+tn+fn)
print(f'After test data analysis the precision is about {pre}\nRecoil is {rec}\nF1 score is {f1}\nAccuracy is {acc} \nAfter 30000 epoches')
```
