1.      Create a class called Matrix with methods to perform various matrix operations using the Numpy library. The class should take two matrices as input attributes and implement the following methods: 

a)      elementwise_addition()

b)      elementwise_subtraction()

c)      elementwise_multiplication()

d)      scalar_multiplication(scalar)

e)      determinant()

f)      matrix_multiplication()

g)      transpose()

h)      inverse()

i)       diagonal(): Print the diagonal elements of the first matrix.

j)       rank()

k)      eigen() : Calculate the eigenvalues and eigenvectors of the first matrix

Note: The eigen() method should also plot the eigenvectors using Matplotlib, which should be imported and used inside the Matrix class.

If the matrices are incompatible for a particular operation or the matrix is not square for calculating the determinant or not invertible for calculating the inverse, a ValueError should be raised

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
class Matrix:
    def __init__(self, m1, m2):
        self.m1 = m1
        self.m2 = m2
    
    def elementwise_additions(self):
        print(np.add(self.m1, self.m2))
        
    def elementwise_subtraction(self):
        print(np.subtract(self.m1, self.m2))
        
    def elementwise_multiplication(self):
        print(np.multiply(self.m1, self.m2))
    
    def scalar_multiplication(self, scalar):
        print(np.multiply(self.m1, scalar))
        
    def determinant(self):
        print('determinant of 1st matrix is: ', np.linalg.det(m1))
    
    def matrix_multiplication(self):
        print(np.dot(self.m1, self.m2))
    
    def transpose(self):
        print(np.array(self.m1).T)
        
    def inverse(self):
        print('inverse of 1st matrix is: ', np.linalg.inv(self.m1))
    
    def diagonal(self):
        print(np.array(self.m1).diagonal())
        
    def rank(self):
        print(np.linalg.matrix_rank(self.m1))
        
    def eigen(self):
        e = np.linalg.eig(self.m1)
        print(f'the eigen values of given matrix are {e[0]} and its eigen vectors are {e[1]}')
        o = np.array([[0, 0], [0, 0]])
        plt.quiver(*o, e[1][:, 0], e[1][:, 1], color=['red','green'], scale=5)
        plt.ylim(-1,1)
        plt.xlim(-1,1)
        EV_Plot=plt.show()
    
```

```python
m1 = [[1, 2], [3, 4]]
m2 = [[5, 6], [2, 5]]
m = Matrix(m1, m2)
print('the element wise addition is ') 
m.elementwise_additions()
print('the element wise substraction is ') 
m.elementwise_subtraction()
print('the element wise multiplication is ') 
m.elementwise_multiplication()
print('the scalar multiplication is ') 
m.scalar_multiplication(0)
print('the detemenant of mat 1 is ') 
m.determinant()
print('the matrix multiplication is ') 
m.matrix_multiplication()
print('the transpose of mat 1 is ') 
m.transpose()
print('the inverse of mat 1 is ') 
m.inverse()
print('the diagonal element of mat 1 is ') 
m.diagonal()
print('the ranl of mat 1 is ') 
m.rank()
m.eigen()
```

```python
# 4
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3d curve')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

3.      Write the Python code to generate contour plots for three different sigmoid functions.  The three sigmoid functions are:

f(x, y) = 1 / (1 + exp(-(x + y + 1)))

f(x, y) = 1 / (1 + exp(-(x + y + xy)))

f(x, y) = 1 / (1 + exp(-((x2) + (y2) + (xy))))

The output should be three separate contour plots, one for each function.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)

Z1 = 1 / (1 + np.exp(-(X + Y + 1)))
Z2 = 1 / (1 + np.exp(-(X + Y + X*Y)))
Z3 =  1 / (1 + np.exp(-((X**2) + (Y**2) + (X*Y))))

fig, axs = plt.subplots(1, 3, figsize=(12, 4))

axs[0].contourf(X, Y, Z1, levels=20)
axs[0].set_title('sigmoid1')

axs[1].contourf(X, Y, Z2, levels=20)
axs[1].set_title('sigmoid2')

axs[2].contourf(X, Y, Z3, levels=20)
axs[2].set_title('sigmoid3')

plt.show()
```

4.   CSV File input output   

a. Write a python program to create a text file in CSV format containing two columns - "Product Name" and "Price". Fill in at least 10 rows of data for different products and their corresponding prices. ( You can give the values in the program itself)

b. Do Question 4a using CSV reader

c. Do Question 4a with Pandas again



```python
import csv

# Define some sample data
product_data = [
    ['Product 1', 10.99],
    ['Product 2', 20.99],
    ['Product 3', 5.99],
    ['Product 4', 15.99],
    ['Product 5', 12.99],
    ['Product 6', 8.99],
    ['Product 7', 18.99],
    ['Product 8', 9.99],
    ['Product 9', 6.99],
    ['Product 10', 13.99],
]


# Define the CSV file name and headers
file_name = 'product_prices.csv'
headers = ['Product Name', 'Price']

with open(file_name, 'w') as file:
    file.write(headers[0])
    file.write(',')
    file.write(headers[1])
    file.write('\n')   
    for row in product_data:
        file.write(row[0])
        file.write(',')
        file.write(str(row[1]))
        file.write('\n')   
        
        
```

```python
#B

# Define the CSV file name
file_name = 'product_prices.csv'

# Open the CSV file for reading
with open(file_name, 'r') as file:
    # Create a CSV reader object
    reader = csv.reader(file)

    # Read the headers from the file
    headers = next(reader)

    # Initialize a dictionary to store the product data
    product_data = {}

    # Read the remaining rows from the file and store them in the dictionary
    for row in reader:
        product_name, price = row
        product_data[product_name] = float(price)

# Print the product data
print("Product Name\tPrice")
for product_name, price in product_data.items():
    print(f"{product_name}\t{price}")
```

Create a .txt (text) file and use the pledge of India as the content of the text file.

Write a python program that reads this text file, processes it by counting the number of occurrences of each word in the file, and then writes the result back to a new text file.

```python
file = open('pledge.txt', 'w')
file.write('India is my country and all Indians are my brothers and sisters. I love my country and I am proud of its rich and varied heritage. I shall always strive to be worthy of it. I shall give respect to my parents, teachers and elders and treat everyone with courtesy.')
```

```python
file.close()
```

```python
with open('wordOccurance.txt', 'w') as f:
    f.write('Word ---- Count\n')
```

```python
with open('pledge.txt', 'r') as file:
    l = file.read().split()
    for item in set(l):
        with open('wordOccurance.txt', 'a') as f:
            f.write(item)
            f.write(' ---- ')
            f.write(str(l.count(item)))
            f.write('\n')
```
