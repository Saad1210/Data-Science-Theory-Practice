```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5, 5, 100)
y = x**2
plt.plot(x, y, color='y')
```

```python
X = [2, 4, 6, 8, 10]
Y = [1, 4, 9, 19, 39]

ax.plot(X, Y)

ax.set(xlabel='x', ylabel='y', title='Test plot')
plt.show()
```

```python
x = np.arange(-2, 2, 0.01)
y = x**2

fig, ax = plt.subplots(2, 2, figsize = (4, 4))

ax[0][0].plot(x, y)
ax[0][1].plot(x, y**0.5)
ax[1][0].plot(X, Y)
ax[1][1].plot(x, y**-1)
```

```python
t = np.arange(-2, 2, 0.1)
x = t
y = t**2
z = t**3
# fig, ax = plt.subplots(figsize = (4, 4))
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot(x, y, z)
```

```python
X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 2, 0.1)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X, Y, Z)
```
